# Caso de Estudio: La encuesta
La alcaldía de la ciudad desea una aplicación que permita realizar una encuesta de opinión sobre el manejo de la crisis generada por el COVID-19.   La encuesta consiste en una única pregunta, en la cual se le pide a la persona dar su opinión acerca del manejo, dando un valor entre 0 y 10 (o es malo y 10 muy bueno).

</br>

Se desea poder conocer los resultados de la encuesta para todos los encuestados y tambien por rango de edad. (mayores de edad y menores de edad) 

El programa debe permitir: 
1. Visualizar el promedio total de la encuesta. 
2. Visualizar el número total de opiniones recogidas con la encuesta. 
3. Agregar una opinión a la encuesta especificando si es mayor o menor de edad 
4. Consultar el valor de la encuesta para un rango de edad y estado civil (personas y promedio).

### ¿Quién es el usuario?
Funcionarios de la Alcaldía

## Definir el problema 
¿Con cuál información contamos?

+ olo hay una pregunta en la encuesta que se califica de 0 a 10
+  encuesta esta dividida en rangos de edades (menores y mayores de edad)


¿Qué es lo que el sistema debe calcular?
+ Número total de opiniones recogidas con la encuesta.
+ Promedio total de la encuesta.  (¿Como se calcula un promedio?)
+ Valor de la encuesta para mayores y para menores de edad.
